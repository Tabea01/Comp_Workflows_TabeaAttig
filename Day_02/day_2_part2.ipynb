Explanation of each summary statistic:
1. Per base sequence quality: A sequence quality score of 30 or above in the the results of sequencing quality per base is considered good. For Sham_oxy_1_1.gz, ... the average sequence quality per base is within this range. 

2. Per tile sequence quality:
The quality of each tile sequenced can be determined through a color scale. A tile is considered good, when it consists of cooler colors indicating higher quality. Hotter colors indicate lower quality. 

3. Per sequence quality scores: 90% of the reads with a quality score of 35 or above are considered to have good quality. If we have lots of low quality scores, there could be a problem with the sequencing run itself. 

4. Per base sequence content: In a high-quality sequencing sample, the four lines indicating the base proportions at each position should be nearly parallel and closely aligned

5. Per sequence GC content: The GC content distribution is analyzed by comparing the observed distribution to a reference normal distribution, helping identify any biases or contamination in the sequencing library; deviations greater than 15% are flagged as warnings, while those exceeding 30% are considered errors.

6. Per base N content: The N ratio per base indicates the percentage of indeterminate bases in sequencing reads, and if it exceeds 5%, it serves as a warning of potential sequencing issues, while values over 20% are considered errors, warranting investigation to maintain data quality.

7. Sequence Length Distribution: A uniform sequence length distribution indicates all sequences in the library have the same length, which is typical for platforms like Illumina; non-uniform lengths may require trimming or specialized tools for downstream analyses, and reads of length 0 signal errors that must be addressed.

8. Sequence Duplication Levels: Repeated sequences measure duplication or bias in sequencing data; while some duplication is expected with high sequencing depth, over 20% duplication may be a warning, and exceeding 50% signals a significant issue that must be corrected to ensure accurate downstream analysis.

9. Overrepresented Sequences: An over-represented sequence is one that occurs frequently in sequencing reads, with a threshold of 0.1% for warning and 1% for an error, assessed on the first 200,000 reads to detect potential contaminants or biases.

10. Adapter Content: Adapter sequences are short DNA fragments attached to the ends of DNA before sequencing, and their presence is tracked as a cumulative percentage to assess how frequently they appear at different positions in the sequenced DNA fragments.
